```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# read all data
dataset = pd.read_csv('dataset1.csv')
dataset
```

```python
dataset.dtypes
```

```python
#Laporan jurusan yang diambil di universitas

apply_Topic = dataset['Prestasi'].value_counts().reset_index()
apply_Topic.columns = ['Prestasi', 'Jumlah']
apply_Topic
```

```python

plt.figure(figsize=(12,8))
plt.bar(apply_Topic['Prestasi'], apply_Topic['Jumlah'])
plt.xlabel('Prestasi')
plt.ylabel('Jumlah')
plt.show()
```

```python
#Laporan jurusan yang diambil di universitas

apply_Topic2 = dataset['Prestasi'].value_counts().reset_index()
apply_Topic2.columns = ['Prestasi', 'Angkatan']
apply_Topic2
```

```python
plt.figure(figsize=(12,8))
plt.bar(apply_Topic2['Prestasi'], apply_Topic2['Angkatan'])
plt.xlabel('Prestasi')
plt.ylabel('Angkatan')
plt.show()
```
